```python
# function:
def distance(user1ID, user2ID):
    user1rating = RatingMatrix.transpose()[user1ID]
    user2rating = RatingMatrix.transpose()[user2ID]
    
    commen_movie = []

    for column in RatingMatrix.columns:
        if user1rating[column] > 0 and user2rating[column] > 0:
            commen_movie.append(column)
        
    user1 = []
    user2 = []
    
    for i in commen_movie:
        user1.append(user1rating[i])
        user2.append(user2rating[i])
        
    # distance:
    if(len(commen_movie)==0):
        distance =-1
    else:
        distance = euclidean(user1, user2)
    
    return distance
```

# part 4: K Nearest Neighbors

```python
target_user = 100
```

```python
commen_user = [] 
distance_to_target = []

for user in RatingMatrix.index:
    if(distance(target_user, user)!=-1):
        commen_user.append(user)
        distance_to_target.append(distance(target_user, user))
        
userDataFrame = pd.DataFrame(distance_to_target, commen_user, columns=["distance"])
userDataFrame = userDataFrame[userDataFrame.index!=target_user]
```

```python
userDataFrame
```

```python
k=10
KNearestNeighbor = userDataFrame.sort_values(["distance"], ascending=True)[:k]
```

```python
KNearestNeighbor
```

```python
def kNearestNeighbor(user, k=10):
    commen_user = [] 
    distance_to_target = []

    for user in RatingMatrix.index:
        if(distance(target_user, user)!=-1):
            commen_user.append(user)
            distance_to_target.append(distance(target_user, user))
        
    userDataFrame = pd.DataFrame(distance_to_target, commen_user, columns=["distance"])
    KNearestNeighbor = userDataFrame.sort_values(["distance"], ascending=True)[:k]
    
    return KNearestNeighbor
```

```python
KNN=kNearestNeighbor(20)
```

```python
def TopRecomended(user, N=10):
    RecommendedMovies = []
    
    # get K Nearest Neighbor Ratings: 
    KNNRatings = RatingMatrix[RatingMatrix.index.isin(KNN.index)]
    
    # get Avrage of Ratings:
    avgRatign = KNNRatings.apply(np.nanmean).dropna()
    
    # get movies that target user Already Seen:
    MovieAlreadySeen = RatingMatrix.transpose()[user].dropna().index
    avgRatign = avgRatign[~avgRatign.isin(MovieAlreadySeen)]
    
    # find top N Movies to Target User:
    TopMovies = avgRatign.sort_values(ascending=False).index[:N]
    
    # Recommended Top N Movies to Target User:
    for movie in TopMovies:
        RecommendedMovies.append(movies["title"][movie])

    return RecommendedMovies
```
